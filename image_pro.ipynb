```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
def histogram_equalize(img):
    b, g, r = cv2.split(img)
    red = cv2.equalizeHist(r)
    green = cv2.equalizeHist(g)
    blue = cv2.equalizeHist(b)
    return cv2.merge((blue, green, red))
kernel = np.ones((5,5),np.uint8)

def histogram_equalize_r(img):
    b, g, r = cv2.split(img)
    red = cv2.equalizeHist(r)
    green = cv2.equalizeHist(g)
    blue = cv2.equalizeHist(b)
    return cv2.merge((blue, green, red))
```

```python
for i in os.listdir("/media/abhik/Windows/ML/data/viz/yes_v"):
    img = cv2.imread("/media/abhik/Windows/ML/data/viz/yes_v/" + str(i))
    step1 = histogram_equalize(img)
    lab= cv2.cvtColor(step1, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(10,10))
    cl = clahe.apply(l)
    print(cl.shape)
```
